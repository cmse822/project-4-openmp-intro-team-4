```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./matrix_multultiplication_Part3_final.csv')
```

```python
matrix_sizes = [64, 128, 256, 512, 1024, 2048]
ranks = [2,4,8,16]
threads = np.arange(1, 13) * 2
```

```python
for i in range(len(matrix_sizes)):
    data_size = data[data["Matrix Size"] == matrix_sizes[i]]

    for rank in ranks:
        data_rank = data_size[data_size['MPI Tasks'] == rank]
        plt.scatter(data_rank['OpenMP Threads'], data_rank['Average Runtime'], label = f'Rank = {rank}')
        
    plt.legend()
    plt.xlabel('Number of threads')
    plt.ylabel('Time (s)')
    plt.title(f'Matrix size = {matrix_sizes[i]} performance')
    plt.show()
```

```python
for i in range(len(ranks)):
    data_rank = data[data["MPI Tasks"] == ranks[i]]

    for thread in threads:
        data_threads = data_rank[data_rank['OpenMP Threads'] == thread]
        plt.scatter(data_threads['Matrix Size'], data_threads['Average Runtime'], label = f'Threads = {thread}')
        
    plt.legend()
    plt.xlabel('Matrix Size')
    plt.ylabel('Time (s)')
    plt.title(f'Rank = {ranks[i]} performance')
    plt.show()
```

```python
for i in range(len(threads)):
    data_thread = data[data["OpenMP Threads"] == threads[i]]

    for rank in ranks:
        data_ranks = data_thread[data_thread['MPI Tasks'] == rank]
        plt.scatter(data_ranks['Matrix Size'], data_ranks['Average Runtime'], label = f'Ranks = {rank}')
        
    plt.legend()
    plt.xlabel('Matrix Size')
    plt.ylabel('Time (s)')
    plt.title(f'Thread = {threads[i]} performance')
    plt.show()
```
